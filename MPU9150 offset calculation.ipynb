# MPU calibration

There are several cells below. I noticed that my MPU9150 has a huge offset in Z axis, around 0.4 m/s2. First I measure some axis using iPhone level app as reference. When object is on equilibrium over a table for example with Z perpendicular to the table, should measure 0,0,+-1 being X,Y,Z and acceleration measured on g (9.8ms2)

```python
#X+ Axis calibration


import serial
import time
import csv

import pandas as pd
import numpy as np



ser = serial.Serial('COM3',921600,timeout=50000)
ser.flushInput()
with open("calx+.csv","a") as f:
    f.write("sep=;\n")
    f.write("timestamp;AX;AY;AZ;GX;GY;GZ;MX;MY;MZ;Roll;Pitch;Yaw;Q1;Q2;Q3;Q4;LABEL\n")
    for i in range(100):
        try:
            ser_bytes = ser.readline()
            decoded_bytes = ser_bytes[0:len(ser_bytes)-2].decode("ASCII")
            #splitted = decoded_bytes.split(";")
            #print(splitted)
            
            #fieldnames = ['timestamp','AX','AY','AZ','GX','GY','GZ','MX','MY','MZ','Roll','Pitch','Yaw','Q1','Q2','Q3','Q4']
            #writer = csv.writer(f,delimiter=';')
            f.write(str(time.time())+";"+decoded_bytes+";FREE"+'\n')
            #writer.writerow(decoded_bytes)
            
        except:
            print("Keyboard Interrupt")
            f.close() 
            ser.close()


f.close() 
ser.close()
            
#header  ['timestamp','AX','AY','AZ','GX','GY','GZ','MX','MY','MZ','Roll','Pitch','Yaw','Q1','Q2','Q3','Q4']

calibration_y_dataset = pd.read_csv('calx+.csv',skiprows=1,delimiter=';')

calibration_y_dataset.head(10)

xposcal =(calibration_y_dataset['AX'] - 9.8 ).mean()
```

```python
#Z Axis calibration +

import serial
import time
import csv

import pandas as pd
import numpy as np



ser = serial.Serial('COM18',921600,timeout=50000)
ser.flushInput()
with open("calz+.csv","a") as f:
    f.write("sep=;\n")
    f.write("timestamp;AX;AY;AZ;GX;GY;GZ;MX;MY;MZ;Roll;Pitch;Yaw;Q1;Q2;Q3;Q4;LABEL\n")
    for i in range(100):
        try:
            ser_bytes = ser.readline()
            decoded_bytes = ser_bytes[0:len(ser_bytes)-2].decode("ASCII")
            #splitted = decoded_bytes.split(";")
            #print(splitted)
            
            #fieldnames = ['timestamp','AX','AY','AZ','GX','GY','GZ','MX','MY','MZ','Roll','Pitch','Yaw','Q1','Q2','Q3','Q4']
            #writer = csv.writer(f,delimiter=';')
            f.write(str(time.time())+";"+decoded_bytes+";FREE"+'\n')
            #writer.writerow(decoded_bytes)
            
        except:
            print("Keyboard Interrupt")
            f.close() 
            ser.close()


f.close() 
ser.close()
            
header = ['timestamp','AX','AY','AZ','GX','GY','GZ','MX','MY','MZ','Roll','Pitch','Yaw','Q1','Q2','Q3','Q4']

calibration_z_dataset = pd.read_csv('calz+.csv',skiprows=1,delimiter=';')

calibration_z_dataset.head(10)

(calibration_z_dataset['AZ'] - 9.8 ).mean()
```
